class mirip seperti mendeskripsikan sebuah objek

```python
class Manusia:
    def __init__(self, warna_rambut, tinggi=150, berat=60): # inisialisasi self = diri sendiri
        self.warna_rambut = warna_rambut
        self.tinggi = tinggi
        self.berat = berat
        
    def cat_rambutnya(self, warna): # Method / action
        self.warna_rambut = warna
        print(f"Rambutnya jadi warna {warna}")
    
    def BMI(self):
        return self.berat / self.tinggi ** 2
```

```python
wira = Manusia("Hitam", 178, 60)
```

## Inheritance 

```python
class Mahasiswa(Manusia):
    def __init__(self, warna_rambut, tinggi, berat, univ):
        super().__init__(warna_rambut, tinggi, berat)
        self.universitas = univ
        
    def pindah_univ(self, univ_baru):
        self.universitas = univ_baru
```

```python
terry = Mahasiswa("hitam", 150, 45, "ITB")
```

```python
terry.universitas
```

```python
terry.pindah_univ("MIT")
```

```python
terry.universitas
```

#### Forward


```python
class CustomNN(nn.Module):
    def __init__(self, input_size, output_size):
        super().__init__()
        self.fc = nn.Sequential(
            nn.Linear(input_size, 16),
            nn.ReLU(),
            nn.Linear(16, 8),
            nn.ReLU(),
            nn.Linear(8, output_size),
            nn.Sigmoid()
        ) # fully conected = Sequential nn
        
    def forward(self, x): # x adalah input data. code backprop sudah di code di balik layar, tapi d feed forward. x masukin fc keluar output
        return self.fc(x)
```

Multitask learning tidak bisa linear

```python
class FaceBiometric(nn.Module):
    def __init__(self, input_size, output_size):
        super().__init__()
        self.fc = nn.Sequential(
            nn.Linear(input_size, 16),
            nn.ReLU()
        )
        
        self.reg = nn.Linear(16, 4)
        self.clf = nn.Sequentail(
            nn.Linear(16, 4),
            nn.Sigmoid()
        )
        
    def forward(self, x):
        x = self.fc(x)
        pred_reg = self.reg(x)
        pred_clf = self.clf(x)
        return pred_reg, pred_clf
```

```python
model = CustomNN(7, 1)
```

```python
contoh_data = torch.rand(100, 7)
```

```python
model.forward(contoh_data)
```

```python
model(contoh_data)
```
