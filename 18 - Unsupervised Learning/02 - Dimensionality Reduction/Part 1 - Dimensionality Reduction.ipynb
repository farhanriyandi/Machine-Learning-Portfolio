# Bagaimana mencari Principal Component

Dimensi baru (garis hijau) itu disebut sebagai principle component.<br>
Component tersebut ditemukan dengan mencari bidang yang orthogonal.<br>
Secara matematik, kita dapat memperoleh bidang orthogonal menggunakan SVD (Singular Value Decomposition)

# PCA with scikit-learn

```python
from sklearn.decomposition import PCA
```

```python
X = demo.load_sample_data()
X.shape
```

```python
plt.figure(figsize=(5, 5))
plt.scatter(X[:, 0], X[:, 1], s=20, c='b');
```

```python
pca = PCA(n_components=1)
X_pca = pca.fit_transform(X)
X_pca.shape
```

```python
pca.components_
```

```python
# mengembalikan ke dua dimensi
X_inv = pca.inverse_transform(X_pca)
X_inv.shape
```

# Visualisasi PCA

```python
plt.figure(figsize=(16, 5))

plt.subplot(131)
plt.scatter(X[:, 0], X[:, 1], s=20, c='b');
plt.title("$X$", fontsize=14)

plt.subplot(132)
plt.scatter(X_pca, np.zeros_like(X_pca), s=20, c='b');
plt.yticks([])
plt.title("$X_{pca}$", fontsize=14);

plt.subplot(133)
plt.scatter(X_inv[:, 0], X_inv[:, 1], s=20, c='b')
plt.title("$X_{inv}$", fontsize=14);
```
