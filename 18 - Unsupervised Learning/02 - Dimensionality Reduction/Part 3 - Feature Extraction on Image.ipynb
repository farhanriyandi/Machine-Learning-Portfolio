```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

from sklearn.decomposition import PCA, NMF

from luwiji.dimensionality_reduction import illustration
```

```python
df = pd.read_csv("data/lfw_people.csv")
df.head()
```

# Dataset Splitting

```python
X = df.drop(columns="name").values
y = df.name.values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

# Visualize Data

```python
idx = 5
plt.imshow(X_train[idx].reshape(87, 65), cmap='gray')
plt.title(y_train[idx])
plt.axis("off")
```

# Decide n_components

```python
pca = PCA().fit(X_train)
plt.plot(pca.explained_variance_ratio_.cumsum())
plt.xlabel("n_components")
plt.ylabel("cumulative Explained Variance");
```

# Feature Extraction melalui Dim Reduction

```python
pca = PCA(250, whiten=True)
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)
```

```python
print(X_train.shape) # hidden layer
print(X_train_pca.shape) # weight / code
print(pca.components_.shape) # fitur
```

## coba visualisasi fitur (component) nya 

```python
plt.figure(figsize=(15, 8))
for i in range(10):
    plt.subplot(2, 5, i+1)
    plt.imshow(pca.components_[i].reshape(87, 65), cmap='gray')
    plt.title(f"PCA komponen ke-{i+1}")
    plt.axis("off");
```

# Reconstruct Image

```python
X_train_inv = pca.inverse_transform(X_train_pca)
```

```python
plt.figure(figsize=(8, 5))

plt.subplot(121)
plt.imshow(X_train[5].reshape(87, 65), cmap='gray')
plt.title(y_train[5])
plt.axis('off');

plt.subplot(122)
plt.imshow(X_train_inv[5].reshape(87, 65), cmap='gray')
plt.title(y_train[5])
plt.axis('off');
```

# Other Alternative -> Non-Negative  Matrix Factorization

- SVD mencari bidang orthogonal
- NMF faktorisasi matrix tapi menjamin komponennya non negative

```python
nmf = NMF(250)
X_train_nmf = nmf.fit_transform(X_train)
X_test_nmf = nmf.transform(X_test)

plt.figure(figsize=(15, 12))
for i in range(15):
    plt.subplot(3, 5, i+1)
    plt.imshow(nmf.components_[i].reshape(87, 65), cmap='gray')
    plt.title(f"NMF componen ke-{i+1}")
    plt.axis("off");
```

```python
#encode
s1 = nmf.transform((soekarno1*255).astype(int))
s2 = nmf.transform((soekarno2*255).astype(int))
h1 = nmf.transform((hatta*255).astype(int))
```

```python
cosine_similarity(s1, s2)
```

```python
cosine_similarity(s1, h1)
```
