# Regression Metrics

### R-squared

Seberapa bagus model kita dibandingkan rata-rata semua. Berkisar dari -inf sampai 1.<br>
Negatif artinya lebih buruk dari menggunakan rata-rata sederhana.<br>
0 artinya sama baiknya dengan rata-rata.
Positif artinya lebih baik dari rata-rata, dan 1 artinya error = 0

### Mean Absolute Error (MAE)

Rata-rata dari error yang dimutlakkan. Semakin kecil semakin baik.

y = data asli,y_topi = data prediksi

### Mean Squared Error (MSE)

Error terhadap data kita kuadratkan lalu dirata-ratakan. Semakin kecil semakin baik.

### Root Mean Squared Error (RMSE)

akar dari MSE
